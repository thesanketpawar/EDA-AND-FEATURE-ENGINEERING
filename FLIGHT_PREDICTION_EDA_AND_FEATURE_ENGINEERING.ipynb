```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_excel('/content/drive/MyDrive/Flight Prediction/Data_Train.xlsx')
```

```python
train_df.head()
```

```python
test_df=pd.read_excel('/content/drive/MyDrive/Flight Prediction/Test_set.xlsx')
```

```python
test_df.head()
```

```python
#Appending the dataset
final_df=train_df.append(test_df)
final_df
```

```python
final_df.info()
```

```python
#Feature Engineering Process on Date_of_Journey
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0].astype(int)
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1].astype(int)
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
final_df.head(1)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Arrival_Time']= final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Time'].str.split(':').str[0].astype(int)
final_df['Arrival_Minutes']=final_df['Arrival_Time'].str.split(':').str[1].astype(int)
```

```python
final_df.head(1)
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Departure_Hour']=final_df['Dep_Time'].str.split(':').str[0].astype(int)
final_df['Departure_Minutes']=final_df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4,'nan':1})
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df.head(1)
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Duration_hour']=final_df['Duration'].str[0].str.split('h').str[0]
```

```python
final_df['Duration_hour']=final_df['Duration_hour'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Duration_hour']=final_df['Duration_hour']*60
```

```python
final_df.head()
```

```python
final_df['Duration_minute']=final_df['Duration'].str[3:].str.split('m').str[0]
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Airline'].unique()
```

```python
#Labeling And Encoding
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
```

```python
final_df.head(1)
```

```python
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
```

```python
final_df.head()
```

```python
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```

```python
final_df.info()
```
